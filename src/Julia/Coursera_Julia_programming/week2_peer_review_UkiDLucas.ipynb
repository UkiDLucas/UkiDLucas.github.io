## Part 3 loop

- Create a new code cell below. 
- In the code cell write a 'For' loop to print x and y values for you function over a range such as -10 to 10. 
- The range that you choose will depend on your function choice. 
- For clarity use a small number of loop iterations to illustrate some of the data point values. 
- Execute the code to confirm in runs and displays the expected values.

```julia
x_array = collect(-5:5) # define a representative range for x
```

```julia
y_array = Array{Float64, 1}(undef, length(x_array)) # declare an array for y float values, undelined values
index = 1 # declare a counter (index)
```

```julia
for x in x_array
    y_array[index] = simpleMath(x)
    index = index + 1
end

y_array # show resulting array y.
```

## Part 4 plot

- Create a new code cell below. 
- In the code cell write a 'plot' statement to **display your function visually**.

```julia
using Plots # https://docs.juliaplots.org/latest/attributes/
gr() # Activate the GR backend for use with Plots https://github.com/jheinen/GR.jl
plot(
    x_array, 
    y_array, 
    fillrange=-4, 
    fillalpha=0.1,
    fillcolor=:blue,
    label   = ("y values"),
    legend  = :top,
    marker  = ([ :star4 ], 3) # type and size
    ) 
```
