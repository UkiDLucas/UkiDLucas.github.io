# My Equation

```julia
f(x) = 50*cos(x) + 3*x^2
```

f is the sum of a sinusoidal function and a quadratic function

```julia
for x = -10:10
    y = f(x)
    println("x = $x, y = $y")
end
```

```julia
using Plots
plotly()

plot(f, -10, 10, 
title = "f(x) = 50*cos(x) + 3*x^2",
xlabel = "x", 
ylabel = "f(x)", 
legend = false, 
color = "green", 
lw = 3)
```
