```julia
x = collect(1:7)
```

```julia
plot!(
    x, y.(x), 
    marker = :diamond,  # markers at the data points
    linewidth=2)        # thicker line
```

        - **Correct** then to a thicker line
        - **Correct** then adds the title
        - **Correct** and finally it removes the legend
